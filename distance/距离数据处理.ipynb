```python
import pandas as pd
import numpy as np 
```

```python
parcel_distance=pd.read_csv('./parcel_distance.csv')
distance_matrix = parcel_distance.pivot(index='parcel_idO', columns='parcel_idD', values='distance')
# 修改index和columns的名字
distance_matrix.index.name='origin'
distance_matrix.columns.name='parcel_id'
# 数值全除以1000
distance_matrix=distance_matrix/1000
distance_matrix.index =  distance_matrix.index + 1
distance_matrix.columns =  distance_matrix.columns + 1
distance_matrix
```
